```python
# Step 2: Load the Countries - US,IN,GB,JP and add a new column region.
```

```python
import pandas as pd
import os

folder_path = "C:/Users/user/Desktop/Youtube project datasets"
region_files = ['US_videos.csv','IN_videos.csv','GB_videos.csv','JP_videos.csv']

dataframes = []

for file in region_files:
    region = file[:2]  # 'US', 'IN', etc.
    df = pd.read_csv(os.path.join(folder_path, file))
    df['region'] = region
    dataframes.append(df)

data = pd.concat(dataframes, ignore_index=True)
data.tail()
```

```python
# Step 3: Convert Dates to datetime Format and Calculate Trending Duration
```

```python
data['trending_date'].head(10)
```

```python
data['trending_date'] = pd.to_datetime(data['trending_date'], errors='coerce')
data['trending_date'].head(10)
```

```python
data['publishedAt'] = pd.to_datetime(data['publishedAt'],errors='coerce')
data['publishedAt'].head(10)
```

```python
# Trending Duration
data['trending_duration'] = (data['trending_date'] - data['publishedAt']).dt.days
data[['publishedAt', 'trending_date', 'trending_duration']].tail()
```

```python
# STEP 4: Map category_id to Category Name
import json
# Load JSON file
with open("C:/Users/user/Desktop/Youtube project datasets/US_category_id.json", 'r') as f:
    category_data = json.load(f)
```

```python
category_mapping = {}

for item in category_data['items']:
    category_mapping[int(item['id'])] = item['snippet']['title']
```

```python
data['category_name'] = data['categoryId'].map(category_mapping)
data.head()
```

```python
# Step 5: DATA CLEANING
```

```python
data.isnull().sum()
```

```python
# remove null, fill nulls
data = data[data['channelTitle'].notnull()]
data['description'] = data['description'].fillna('')
```

```python
# Check duplicate rows (based on video_id, trending_date, and region) and drop them
duplicates = data.duplicated(subset=['video_id', 'trending_date', 'region'])
data = data.drop_duplicates(subset=['video_id', 'trending_date', 'region'])
```

```python
# Strip whitespace and lowercase key text fields
data['title'] = data['title'].str.strip().str.lower()
data['tags']= data['tags'].str.strip().str.lower()
data['description']=data['description'].str.strip().str.lower()
```

```python
# Replace '[none]' with empty string
data['tags'] = data['tags'].replace('[none]', '')
```

```python
print("Nulls after cleaning:")
print(data.isnull().sum())
print("Duplicates remaining:",duplicates.sum())
```

```python
from textblob import TextBlob
#function to get sentiment polarity btw -1.0 to 1.0 for title and tags
def get_sentiment(text):
    blob=TextBlob(text)
    return blob.sentiment.polarity
```

```python
#!pip install tqdm
```

```python
from tqdm import tqdm
tqdm.pandas()

data['title_sentiment'] = data['title'].progress_apply(get_sentiment)
data['tags_sentiment']=data['tags'].progress_apply(get_sentiment)
```

```python
def classify_sentiment(score):
    if score > 0.1:
        return 'Positive'
    elif score < -0.1:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
data['title_sentiment_label']=data['title_sentiment'].progress_apply(classify_sentiment)
data['tags_sentiment_label']=data['tags_sentiment'].progress_apply(classify_sentiment)
```

```python
# Step 7 : SQLite to run queries
```

```python
import sqlite3
conn = sqlite3.connect('youtube_trending.db')

data.to_sql('videos', conn, if_exists='replace', index=False)
```

```python
# average views by category
query1 = '''
select category_name, avg(view_count) as avg_views
from videos
group by category_name
order by avg_views desc
'''
avg_views_category = pd.read_sql_query(query1, conn)
avg_views_category.head()
```

```python
# average views by category per region
query2 = '''
select region,category_name,avg(view_count) as avg_views
from videos
group by region,category_name
order by region,avg_views desc
'''
avg_views_region= pd.read_sql_query(query2,conn)
avg_views_region.head()
```

```python
# Step 8 : Visualizations using matplotlib & seaborn
```

```python
#  Time series of trending duration over time per region

data['trending_date'] = pd.to_datetime(data['trending_date'])
data['publishedAt'] = pd.to_datetime(data['publishedAt'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='trending_date', y='trending_duration', hue='region')
plt.title('Trending Duration Over Time by Region')
plt.xlabel('Trending Date')
plt.ylabel('Trending Duration (days)')
plt.tight_layout()
plt.show()
```

```python
# Most Viewed Categories per Region
# Aggregate average views
top_categories = data.groupby(['region', 'category_name'])['view_count'].mean().reset_index()
top_categories = top_categories.sort_values(['region', 'view_count'], ascending=[True, False])

plt.figure(figsize=(14, 8))
sns.barplot(data=top_categories, x='view_count', y='category_name', hue='region')
plt.title('Most Viewed Categories per Region')
plt.xlabel('Average Views')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# Sentiment Scores by Region
# Violin plot: sentiment distribution of titles by region
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='region', y='title_sentiment')
plt.title('Title Sentiment Distribution by Region')
plt.ylabel('Sentiment Polarity')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```
